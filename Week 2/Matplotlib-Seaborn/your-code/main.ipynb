#### Use your previous code but now, resize your plot.
**Hint**: Add the figsize argument in plt.subplots().

If you want, try to add a title to the plot or even axes labels. You can also play with the fontweight and fontsize of the titles and labels. 

```python
# your code here
fig, axes = plt.subplots(1, 2, figsize=(10,5))

axes[0].plot(x, y, 'g--', lw=4)
axes[0].set_title('$y = 2x$', fontsize=24, color='b', pad=20)
axes[0].set_xlabel('x', fontsize=16)
axes[0].set_ylabel('y', fontsize=16)

axes[1].plot(x, z, 'r:', lw=4);
axes[1].set_title('$y = x^2$', fontsize=24, color='b', pad=20)
axes[1].set_xlabel('x', fontsize=16)

fig.tight_layout();
```

#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here
x = np.arange(0,100)
y = x**2
z = np.exp(x)

fig, axes = plt.subplots(1, 2, figsize=(10,5))

axes[0].plot(x, y, 'g--', lw=4)
axes[0].set_title('$y = 2x$', fontsize=24, color='b', pad=20)
axes[0].set_xlabel('x', fontsize=16)
axes[0].set_ylabel('y', fontsize=16)

axes[1].plot(x, z, 'r:', lw=4)
axes[1].set_title('$y = e^x$', fontsize=24, color='b', pad=20)
axes[1].set_xlabel('x', fontsize=16);
axes[1].set_ylabel('log(y)', fontsize=16)
axes[1].set_yscale("log")

fig.tight_layout();
```

#### As a bonus challenge, try to add a legend to the plot.

```python
# your code here
fig, axes = plt.subplots(1, 2, figsize=(10,5))

axes[0].plot(x, y, 'g--', lw=4, label='$2x$')
axes[0].set_xlabel('x', fontsize=16)
axes[0].set_ylabel('y', fontsize=16)
axes[0].legend()

axes[1].plot(x, z, 'r:', lw=4, label='$e^x$')
axes[1].set_xlabel('x', fontsize=16)
axes[1].set_ylabel('log(y)', fontsize=16)
axes[1].set_yscale("log")
axes[1].legend()

fig.tight_layout();
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
df_months = fitbit[['Months_encoded', 'Steps']].groupby(['Months_encoded']).median().reset_index()
fig, ax = plt.subplots(1, 1, figsize=(12, 4))
plt.sca(ax)
sns.barplot(x='Months_encoded', y='Steps', data=df_months)
plt.xticks(range(12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
fig.savefig('../figures/median_setps_by_month_encoded.png')

# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
df_work = fitbit[['Work_or_Weekend', 'Steps']].groupby(['Work_or_Weekend']).median().reset_index()
fig, ax = plt.subplots(1, 1, figsize=(12, 4))
plt.sca(ax)
sns.barplot(x='Work_or_Weekend', y='Steps', data=df_work)
plt.xticks([0, 1], ['Weekend', 'Work'])
fig.savefig('../figures/median_steps_by_work_or_weekend.png');
```

#### Plot the correlation matrix using seaborn.

```python
# your code here
corr_df = titanic.corr()

fig, ax = plt.subplots(figsize=(8,8))

mask = np.zeros_like(corr_df, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
sns.heatmap(corr_df, mask=mask, cmap='Blues', annot=True, square=True, cbar_kws={"shrink": .75})
ax.set_ylim(len(corr_df), 0, 0);
```

#### What are the most correlated features?

The most correlated features are:
- parch and sibsp, with a correlation of 0.41
- alone and adult_male, with a correlation of 0.4
