```python
import shap
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
shap.initjs()
```

# Import the adults dataset

```python
shap.datasets.cache("https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data")
```

```python
X, y = shap.datasets.adult()
```

# Use the sklearn.model_selection.train_test_split to generate a test-dataset containing 20% of the public.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=42)
```

# Train a DecisionTreeClassifier on the train-dataset

```python
tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)
```

# Use the KernelExplainer to start explaining the model

You'll need the function `f` below to pass to the `shap.KernelExplainer`. Change the name `model` to the variable you stored your own model.

**Warning**: use only 10% of your training data, otherwise the results will be computationally intensive.

```python
f = lambda x: tree.predict_proba(x)[:,1]
```

What is this f doing?

```python
# it is outputting the classifiers's predicted probabilities for the target
```

# Select a single row in your dataset and explain that decision. 

Select a single row and use the explainer's method `explainer.shap_values` to calculate the shap value for that row.


```python
explainer = shap.TreeExplainer(tree, X_train, feature_perturbation='tree_path_dependent')
shap_values = explainer.shap_values(X_train.iloc[0])
shap_values[0]
```

# User the `shap.force_plot` explaining method to explore why the model decided for that decision.

```python
shap.force_plot(explainer.expected_value[0], shap_values[0], feature_names=X_test.columns)
```

# Calculate the shap values for 10~20 rows of your test set.

Hint: You can use the nsamples argument to accelerate

```python
X_test_sub = X_test.iloc[:10]
explainer = shap.TreeExplainer(tree, feature_perturbation="tree_path_dependent")
shap_values = explainer.shap_values(X_test_sub)
```

# Use the `shap.summary_plot` to understand some patterns

```python
shap.summary_plot(shap_values, feature_names=X_test.columns)
```

What do you understand from that?

_hint: You can use the argument_ `plot_type = 'violin'` _to see the effects._

```python
shap.summary_plot(shap_values[0], feature_names=X_test.columns, plot_type='violin')
```

# Plot the force plot for these samples you've selected.

```python
shap.force_plot(explainer.expected_value[0],
                shap_values[0],
                feature_names=X_test.columns)
```

What do you understand from this graph?

# Plot a `shap.decision_plot` of those samples. Try to highlight the first row of your sample.

```python
shap.decision_plot(explainer.expected_value[0],
                   shap_values[0],
                   feature_names=list(X.columns),
                   highlight=0,
                   legend_labels=list(X.index),
                   legend_location='lower right')
```
