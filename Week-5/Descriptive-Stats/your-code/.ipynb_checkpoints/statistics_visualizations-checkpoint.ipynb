## Univariate Analysis

#### Gender count:
Count how many Male and Female exist on this dataset. 

```python
# Hint: Another way to graphically see it would be to use: sns.countplot(data['Gender']).
# Also, if you've used value_counts, you can specify normalize=True to check percentages.
data['Gender'].value_counts()
```

### Considering only Height

Create a pandas series of the Height variable

```python
height = pd.Series(data['Height'])
```

```python
height.shape
```

#### Histogram-plot

Plot the histogram of the height

```python
plt.figure(figsize=(12, 8))

height.plot.hist(bins=100);
```

#### Box-plot

Plot the boxplot of the height

```python
plt.figure(figsize=(12, 8))

height.plot.box();
```

What do you think the distribution of height is like? Do you think it is normal for variables to behave like that in real life?

```python
# looks normal
```

#### KDE distribution for height

Plot the kde distribution (without the hist distribution) of the height.

```python
height.plot.density();
```

#### Analysis

As we can see we have a high count for height in the range 60 to 75. How many people fall off 3 standard deviations from the mean? Can you consider them outliers? Why is that?

```python
len(height[(height < height.mean() - 3 * height.std()) | (height > height.mean() + 3 * height.std())])
# no, they are not outliers, since they are not really detached from the main body of the population
```

```python
# expected results:
data[(height < height.mean() - 3 * height.std()) | (height > height.mean() + 3 * height.std())]
```

## Bivariate Analysis

#### Considering both height and weight

Perform a scatterplot to check the relation between Height and Weight

```python
plt.figure(figsize=(12, 8))

sns.scatterplot(x='Height', y='Weight', data=data);
```

Do the same plot, but color the markers by Gender

```python
plt.figure(figsize=(12, 8))

sns.scatterplot(x='Height', y='Weight', hue='Gender', data=data);
```

Create a variable called gender_groupby to group data by 'Gender'. Do not specify any aggregation yet.

```python
gender_groupby = data.groupby('Gender')
```

Run `gender_groupby.head()` to check the groups obtained.

```python
gender_groupby.head()
```

Run gender_groupby.describe() to check the statistics for each group. Do you know how one could enhance the visualization of this result? 

```python
gender_groupby.describe().T
```

### Distribution plots

#### Considering both gender and height

Plot the distribution of the variable Height in the same plot breaking them down by 'Gender'

```python
# Hint: You can use FaceGrid and map from sns or plot twice in the same axis
```

```python
# using Facetgrid and mapsns.FacetGrid(data, col='Gender').map(plt.hist, 'Height', bins=70);
sns.FacetGrid(data, col='Gender', height=8).map(plt.hist, 'Height', bins=70);
```

```python
# plotting sns.distplot twice
fig, ax = plt.subplots(1, 2, figsize=(12, 8))

sns.distplot(data.loc[data['Gender'] == 'Male', 'Height'], bins=70, kde=False, ax=ax[0])
sns.distplot(data.loc[data['Gender'] == 'Female', 'Height'], bins=70, kde=False, ax=ax[1]);
```

#### Considering both gender and weight and gender and money

Do the same for the weight and money variable

```python
# Hint, for the Money variable, try specifying hist=False
sns.FacetGrid(data, col='Gender', height=8).map(plt.hist, 'Weight', bins=70);
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 8))

sns.distplot(data.loc[data['Gender'] == 'Male', 'Money'], bins=70, hist=False, ax=ax[0])
sns.distplot(data.loc[data['Gender'] == 'Female', 'Money'], bins=70, hist=False, ax=ax[1]);
```
